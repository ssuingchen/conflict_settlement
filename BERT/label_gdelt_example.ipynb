```python
from transformers import BertForSequenceClassification, BertTokenizer
import torch
```

```python
# prepare model and tokenizer
model_file="second_finetuned_BERT_epoch_5.model"
model = BertForSequenceClassification.from_pretrained("bert-base-uncased",
                                                      num_labels=11,
                                                      output_attentions=False,
                                                      output_hidden_states=False)
model.load_state_dict(torch.load(model_file, map_location=torch.device('cpu')))
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', do_lower_case=True)
```

Clean GDELT data

```python
gdelt_url_df, gdelt_theme_df = BERT_Classifier.clean_gdelt(gdelt_file="Sierra_Leone_top_companies_wurls.csv")
```

Label GDELT data

```python
BERT_Classifier.label_gdelt(gdelt_url_df, gdelt_theme_df, model, tokenizer)
```
